```python
import pandas as pd
```

```python
data = pd.read_csv('adult_data.csv')
data.head(10)

```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

```python
men = data[ data.sex == 'Male']

women  = data[ data.sex == 'Female']

print('мужчин = %d , женщин = %d' % (len(men.index),len(women.index)))
```

**2. Каков средний возраст (признак *age*) женщин?**

```python
test = data[ data.sex == 'Female'].age.mean()

print('средний возраст женщин = %.2f' % test)
```

**3. Какова доля граждан Германии (признак *native-country*)?**

```python
newdata = data[ data['native-country'] == 'Germany']


print('доля граждан Германии = %.2f%%' % (len(newdata.index)/len(data.index)*100))
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
age_mean_more = data[ data.salary == '>50K'].age.mean()

age_std_more = data[ data.salary == '>50K'].age.std()

age_mean_under = data[ data.salary == '<=50K'].age.mean()

age_std_under = data[ data.salary == '<=50K'].age.std()

print('средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K  = %.2f , %.2f' % (age_mean_more,age_std_more))
print('средние значения и среднеквадратичные отклонения возраста тех, кто получает менее 50K  = %.2f , %.2f' % (age_mean_under,age_std_under))
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
newdata = data.groupby(['race','sex']).age.describe()['max']

age = newdata[ 'Amer-Indian-Eskimo','Male']

print('максимальный возраст мужчин расы Amer-Indian-Eskimo = %.2f' % age)
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python


married = data[ ( data['marital-status'].str.find('Married') == 0) & (data.salary == '>50K') & (data.sex == 'Male')]


unMarried = data[ ( data['marital-status'].str.find('Married') != 0) & (data.salary == '>50K') & (data.sex == 'Male')]


print('доля женатых (>50K) мужчин = %.2f%%' % (len(married.index)/len(data.index)*100))

print('доля холостых (>50K) мужчин = %.2f%%' % (len(unMarried.index)/len(data.index)*100))

print('Среди женатых мужчин больше доля зарабатывающих больше')
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
newdata = data

newdata['counter'] = 1

max_hours = data['hours-per-week'].max()

how_all = data[data['hours-per-week'] == max_hours].counter.count()

how_more_salary = data[ (data['hours-per-week'] == max_hours)  & (data.salary == '>50K') ].counter.count()


print('максимальное число часов человек работает в неделю   = %.2f' %  max_hours  )

print('людей работают такое количество часов = %.2f' %  how_all  )

print('процент зарабатывающих среди них много (>50K) = %.2f%%' % ( how_more_salary/how_all*100)  )
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python


testtemp = data.groupby(['native-country','salary'])['hours-per-week'].mean().reset_index().sort_values(by=['hours-per-week','salary'], ascending=True)



testtemp.head(100)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

testtemp = data.groupby(['native-country','salary'])['hours-per-week'].mean().
reset_index().sort_values(by=['hours-per-week','salary'], ascending=True)

plt.subplots(figsize=(20,20))

plt.axis([0,100,20,70])


plt.xticks(rotation=70)

ax = sns.barplot(data=testtemp,
            x="native-country",
           y="hours-per-week",
          hue="salary")



ax.set(xlabel='Страна',ylabel='Среднее время работы')



leg = ax.get_legend()
leg.set_title('')
new_labels = ['Заработок больше 50K', 'Заработок меньше 50K']
for t, l in zip(leg.texts, new_labels): t.set_text(l)
    
    


plt.show()
```
